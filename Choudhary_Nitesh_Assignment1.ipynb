#####  18) Please filter out the top OBPs for the 1995 season with at least 400 PAs, sorted by OBP.  I would like a dataframe for this

```python
# Assuming
# PA= AB+BB+IBB+SH+SF
# AVG = H/AB
# OBP = (H+BB+IBB+SH+SF)/(AB)
```

```python
data['PA']=data['AB']+data['BB']+data['IBB']+data['SH']+data['SF']
data['OBP']=(data['H']+data['BB']+data['IBB']+data['SH']+data['SF'])/data['AB']
x=data[(data['yearID']==1995) & (data['PA']>400)]
x.sort_values(['OBP'],ascending=False)
```

##### 24) Who is the youngest player to hit 50 or more HRs in a single season?

##### 25) Who are the five youngest players to hit 300 or more HRs for their career?

```python
x=data.groupby(['playerID','nameFirst','nameLast'])['HR'].sum().to_frame('totalHR')
y=data.groupby(['playerID'])['Age'].max().to_frame('age')
```

```python
x.reset_index(inplace=True)
y.reset_index(inplace=True)
xy=x.merge(y,left_on='playerID',right_on='playerID',how='inner')
```

```python
xy[xy['totalHR']>300].sort_values(['age']).head(5)
```

##### Bonus Question 1
##### Graph total HRs per season using bar graph

```python
x=data.groupby(['yearID'])['HR'].sum().to_frame('totalHR').reset_index()
fig,axis=plt.subplots(figsize=(100,150))
axis.set_yticklabels(labels=x['totalHR'],fontsize=10)
sns.barplot(x='yearID',y='totalHR',data=x,orient='h',lw=18)
plt.ylabel('Seasons',fontsize=90)
plt.xlabel('Total HRs',fontsize=90)
```

##### Bonus Question 2
##### Using a line graph please graph the average HRs per AB (think about this) per season

```python
fig,axis=plt.subplots(figsize=(12,6))
axis.plot(x['yearID'],x['avgHRAB'])
plt.xlabel('Seasons',fontsize=15)
plt.ylabel('Avg HRs per AB',fontsize=15)
```

```python
# Regression Curve for average HRs/AB for different seasons
x=data.groupby('yearID')['HR','AB'].sum().reset_index()
x['avgHRAB']=x['HR']/x['AB']
sns.lmplot(y='avgHRAB',x='yearID',data=x,size=10,aspect=4)
plt.ylabel('Avg HRs per AB',fontsize=40)
plt.xlabel('Seasons',fontsize=40)
```
